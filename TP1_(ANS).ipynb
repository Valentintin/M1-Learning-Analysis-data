## le Calcul de centré réduire:

    x_centrée_réduit = (x - u) / s

````python
# Centrer réduire
scaler = sk.preprocessing.StandardScaler()
scaler.fit(X)
X_norm = scaler.transform(X)
````

## PCA 
pca est une méthode qui permet de trouver la meilleur variance des variables.
1. on tire avec :
```python
pca.explained_variance_ratio_
```
la taux d'information conservé par la pca

2. on tire avec :
```python
pca.components_
```
la participation des variables de la donnée dans les nouveaux axes de la pca

le poid de participation est pris en compte si elle respecte la règle empirique : |ax_i| >= 1/sqrt(nombre_de_variable) - 0.02

Quel est le nombre d’axes à retenir pour conserver une bonne quantité de l’information
représentée dans le nuage initial (c.f. attributs de sortie explained_varianace_ et
explained_variance_ratio_ de votre objet pca).

````python
pca = PCA(n_components=2).fit(X_norm)
X_norm_pca = pca.transform(X_norm)  # donnée transformé avec la PCA
print("taux d'information conservé avec 2 dimension de la pca :",sum(pca.explained_variance_ratio_))
print("On voit que pour 2 axes de la pca on conserver plus de 0.98 d'information de la donnée...")
````

• Donner une interprétation des deux premiers axes principaux (c.f. analyser les deux premières
lignes de votre attribut de sortie components de votre objet pca).

````python
def regle_de_seuil(data, pca):
    axes_labels: List = []
    for ind, axes in enumerate(pca.components_):
        print("variables selectionnées pour axes",ind)
        label_res: str = ""
        for mois, poid in zip(data.columns[1:], axes):
            if abs(poid) >= 1/np.sqrt(len(data.columns[1:]) - 0.02):
                print(poid, mois)
                if(poid<0):
                    label_res += '-'+mois+ ", "
                else:
                    label_res += mois+ ", "
        axes_labels.append("description de l'axe "+str(ind)+": "+label_res)
        print("\n")
    return axes_labels
    
xy_labels = regle_de_seuil(data, pca)
````

````python
def plot_pca(X: np.array, labels: np.array, title: str, xy_label):
    
    pca = PCA(n_components=2).fit(X_norm)
    X_norm_pca = pca.transform(X_norm)
    fig, axs = plt.subplots(1, 1)
    axs.scatter(X_norm_pca[:, 0], X_norm_pca[:, 1])
    axs.set_title(title)
    axs.set_xlabel(xy_label[0])
    axs.set_ylabel(xy_label[1])
    for label, x, y in zip(labels, X_norm_pca[:, 0], X_norm_pca[:, 1]):
        axs.annotate(label, xy=(x, y), xytext=(-10, 5.), textcoords='offset points')

plot_pca(X_norm, labels, "Visualisation centrée réduite des villes avec PCA", xy_labels )
````

l'analyse du nuage de points nous montre que l'axe 0 correspond au mois de "transition" entre l'été et l'hiver.
l'axe 1 représente les mois chaud (été) et les mois froid (hiver) donc strasbourg est proche de la moyenne des températures en hiver, mais à écarté de la moyenne temperature en été.

Lyon pour l'axe 0 (printemps et automne), est proche de la moyenne des températures

````python
data = pd.read_csv('./villes.csv', sep=';')
X = data.iloc[:, 1:13].values
labels = data.iloc[:, 0].values
data

# Centrer réduire
scaler = sk.preprocessing.StandardScaler()
scaler.fit(X)
X_norm = scaler.transform(X)
````

1. Appliquez la procédure KMeans sur ce jeu de données pour obtenir 3 clusters
• Donner une visualisation graphique des villes projetées dans le plan principal. Les villes de chaque
cluster devraient avoir une couleur différente des villes des autres clusters (voir code ci-dessous).
X_pca étant la matrice des données transformées par l’ACP, labels étant le vecteur contenant le nom
des instances (ici les villes), clustering étant le clustering obtenu.

````python
k_classes: int = 3
labels_clustering: np.array = cluster.KMeans(k_classes, n_init=10).fit_predict(X_norm)
print(labels_clustering)
def plot_pca_clustering(X: np.array, labels: np.array, title: str, xy_label, clustering):
    
    pca = PCA(n_components=2).fit(X_norm)
    X_norm_pca = pca.transform(X_norm)
    fig, axs = plt.subplots(1, 1)
    colors = ['red','yellow','blue','pink']
    axs.scatter(X_norm_pca[:, 0], X_norm_pca[:, 1], c= clustering, cmap=matplotlib.colors.ListedColormap(colors))
    axs.set_title(title)
    axs.set_xlabel(xy_label[0])
    axs.set_ylabel(xy_label[1])
    for label, x, y in zip(labels, X_norm_pca[:, 0], X_norm_pca[:, 1]):
        axs.annotate(label, xy=(x, y), xytext=(-10, 5), textcoords='offset points')
xy_labels = regle_de_seuil(data, pca)
plot_pca_clustering(X_norm, labels, "Visualisation centrée réduite des villes avec PCA et clustering 3 classes", xy_labels, labels_clustering)
````

2. Appliquez la procédure AgglomerativeClustering sur ce jeu de données pour obtenir trois clusters avec
différentes méthodes d’agrégation (il faut essayer single, ward et average).

````python
single = cluster.AgglomerativeClustering(n_clusters=k_classes, linkage='single').fit_predict(X_norm)

average = cluster.AgglomerativeClustering(n_clusters=k_classes, linkage='average').fit_predict(X_norm)

ward = cluster.AgglomerativeClustering(n_clusters=k_classes, linkage='ward').fit_predict(X_norm)

plot_pca_clustering(X_norm,labels,"Agglomerative clustering, single",xy_labels, single)
plot_pca_clustering(X_norm,labels,"Agglomerative clustering, average",xy_labels, average)
plot_pca_clustering(X_norm,labels,"Agglomerative clustering, ward",xy_labels, ward)
````

3. Utiliser cet indice dans une boucle de 8 itérations au maximum (voir code ci-dessous). Les 8 itérations
correspondent aux 8 partitions possibles i.e. en 2, 3, 4, 5, 6, 7, 8 et 9 classes issues de KMeans. Déduire la
meilleure partition qui correspond à un indice maximum pour l’indice Silhouette et minimum de Davies
Bouldin.

````python
tab_sil = []
tab_bou = []
for i in np.arange(2, 9):
    clustering = cluster.KMeans(n_clusters=i).fit_predict(X_norm)
    sil = sk.metrics.silhouette_score(X_norm, clustering, metric='euclidean')
    tab_sil.append(sil)
    print("pour silhouette",i,"classes :",sil)
    boudin = sk.metrics.davies_bouldin_score(X_norm, clustering)
    tab_bou.append(boudin)
    print("pour david bouldin",i,"classes :",boudin)
    print()
fig, axs = plt.subplots(1, 1)
colors = ['red','yellow','blue','pink']
axs.plot(np.arange(2, 9), tab_sil, label="silhouette")
axs.plot(np.arange(2, 9), tab_bou, label="bouldin")
axs.legend()
axs.set_title("graph des métrics pour silhouette")
axs.set_xlabel("nombre de cluster")
axs.set_ylabel("metric")
````

On voit que le max indice sihouette: 0.62 et le min indice davies_bouldin: 0.51 est pour 2 classes créer par Kmeans...

4. Quelle est le meilleur algorithme à utiliser pour obtenir 3 clusters ? Interprétez le résultat obtenu dans ce
cas.

````python
i=3
kmeans = cluster.KMeans(n_clusters=i).fit_predict(X_norm)
single = cluster.AgglomerativeClustering(n_clusters=i, linkage='single').fit_predict(X_norm)
average = cluster.AgglomerativeClustering(n_clusters=i, linkage='average').fit_predict(X_norm)
ward = cluster.AgglomerativeClustering(n_clusters=i, linkage='ward').fit_predict(X_norm)

kmeans_silhouette = sk.metrics.silhouette_score(X_norm, kmeans, metric='euclidean')
single_silhouette = sk.metrics.silhouette_score(X_norm, single, metric='euclidean')
average_silhouette = sk.metrics.silhouette_score(X_norm, average, metric='euclidean')
ward_silhouette = sk.metrics.silhouette_score(X_norm, ward, metric='euclidean')
print("pour Kmeans",i,"classes :",kmeans_silhouette)
print("pour Agglomerative Clustering single:",i,"classes :",single_silhouette)
print("pour Agglomerative Clustering average:",i,"classes :",average_silhouette)
print("pour Agglomerative Clustering ward:",i,"classes :",ward_silhouette)
print("pour",i,"classes le meilleur clustering est :",max(kmeans_silhouette, single_silhouette, average_silhouette, ward_silhouette))
````

pour 3 classes le meilleur clustering est le average, avec la métric silhouette

````python
def functionFinal(X: np.array):
    
    # Centrer réduire
    scaler = sk.preprocessing.StandardScaler()
    scaler.fit(X)
    X_norm = scaler.transform(X)

    best_metric = -10000
    best_ind = 0
    tab_clust = []
    best_method = ""
    for i in np.arange(2, 10):
        kmeans = cluster.KMeans(n_clusters=i).fit_predict(X_norm)
        single = cluster.AgglomerativeClustering(n_clusters=i, linkage='single').fit_predict(X_norm)
        average = cluster.AgglomerativeClustering(n_clusters=i, linkage='average').fit_predict(X_norm)
        ward = cluster.AgglomerativeClustering(n_clusters=i, linkage='ward').fit_predict(X_norm)
        
        kmeans_silhouette = sk.metrics.silhouette_score(X_norm, kmeans, metric='euclidean')
        best_metric = max(best_metric, kmeans_silhouette)
        single_silhouette = sk.metrics.silhouette_score(X_norm, single, metric='euclidean')
        average_silhouette = sk.metrics.silhouette_score(X_norm, average, metric='euclidean')
        ward_silhouette = sk.metrics.silhouette_score(X_norm, ward, metric='euclidean')
        print("pour Kmeans",i,"classes :",kmeans_silhouette)
        print("pour Agglomerative Clustering single:",i,"classes :",single_silhouette)
        print("pour Agglomerative Clustering average:",i,"classes :",average_silhouette)
        print("pour Agglomerative Clustering ward:",i,"classes :",ward_silhouette)
        max_metric = max(kmeans_silhouette, single_silhouette, average_silhouette, ward_silhouette)
        best_method_tmp = ""
        if kmeans_silhouette == max_metric:
            tab_clust.append(kmeans)
            best_method_tmp = "kmeans"
        if single_silhouette == max_metric:
            tab_clust.append(single)
            best_method_tmp = "single"
        if average_silhouette == max_metric:
            tab_clust.append(average)
            best_method_tmp = "average"
        if ward_silhouette == max_metric:
            tab_clust.append(ward)
            best_method_tmp = "ward"
            
        if max_metric >= best_metric:
            best_ind = i
            best_metric = max_metric
            best_method = best_method_tmp
            
    xy_labels = regle_de_seuil(data, pca)
    plot_pca_clustering(X_norm,labels,"Clustering "+best_method,xy_labels, tab_clust[best_ind])
functionFinal(X)
````
